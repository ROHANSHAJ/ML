```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('YEAR')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'].values.reshape(-1, 1), df['per capita income (US$)'])
plt.show()
```

```python
#TRAIN MODEL {LINEAR REG}
X = df['year'].values.reshape(-1, 1)  #reshape to 2d from 1d array in csv
y = df['per capita income (US$)']
reg=linear_model.LinearRegression()
reg.fit(X, y)
```

```python
reg.predict([[2020]]) #predict data for year 2020
```
